```
import pandas as pd 
```

```
# Load data
df = pd.read_csv("data/health_data.csv")

# Combine symptom columns into one text column
df["symptoms"] = df["symptom_1"].fillna('') + " " + df["symptom_2"].fillna('') + " " + df["symptom_3"].fillna('')

# Drop duplicates / blanks
df.dropna(subset=["disease"], inplace=True)
df.drop_duplicates(subset=["symptoms"], inplace=True)

print("Sample data:\n", df.head())
print("Total records:", len(df))

# Save cleaned data
df.to_csv("data/clean_health_data.csv", index=False)
print("✅ Clean dataset saved as data/clean_health_data.csv")
```
